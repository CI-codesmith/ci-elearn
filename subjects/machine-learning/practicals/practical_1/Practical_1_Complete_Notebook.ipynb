## üì¶ Test 1: Verify Core Library Installations

This test imports all essential ML libraries and displays their versions.

````python
# Test 1: Import essential libraries and check versions
import subprocess

print("\n" + "=" * 60)
print("TEST 1: LIBRARY INSTALLATION VERIFICATION")
print("=" * 60 + "\n")

libraries = {
    'numpy': 'NumPy',
    'pandas': 'Pandas',
    'matplotlib': 'Matplotlib',
    'sklearn': 'Scikit-learn',
    'scipy': 'SciPy',
    'seaborn': 'Seaborn'
}

installed_libraries = {}
failed_libraries = {}

for lib_import, lib_name in libraries.items():
    try:
        lib = __import__(lib_import)
        version = lib.__version__ if hasattr(lib, '__version__') else 'Unknown'
        installed_libraries[lib_name] = version
        print(f"‚úì {lib_name:20s} {version:10s} - INSTALLED")
    except ImportError as e:
        failed_libraries[lib_name] = str(e)
        print(f"‚úó {lib_name:20s} - FAILED TO IMPORT")

print("\n" + "=" * 60)
print(f"Summary: {len(installed_libraries)} installed, {len(failed_libraries)} failed")
print("=" * 60)

if failed_libraries:
    print("\n‚ö†Ô∏è  FAILED IMPORTS:")
    for lib, error in failed_libraries.items():
        print(f"  - {lib}: {error}")
else:
    print("\n‚úì‚úì‚úì ALL LIBRARIES INSTALLED SUCCESSFULLY! ‚úì‚úì‚úì")
````

## üìà Test 4: Matplotlib Visualization

Test Matplotlib for creating visualizations.

````python
import matplotlib.pyplot as plt
import numpy as np

print("\n" + "=" * 60)
print("TEST 4: MATPLOTLIB VISUALIZATION")
print("=" * 60 + "\n")

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Matplotlib Visualization Tests', fontsize=16, fontweight='bold')

# Plot 1: Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0, 0].plot(x, y, 'b-', linewidth=2, label='sin(x)')
axes[0, 0].set_title('Line Plot: sin(x)', fontweight='bold')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].legend()

# Plot 2: Bar chart
categories = ['Math', 'English', 'Science', 'History']
scores = [85, 90, 78, 92]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']
axes[0, 1].bar(categories, scores, color=colors, alpha=0.7, edgecolor='black')
axes[0, 1].set_title('Bar Chart: Subject Scores', fontweight='bold')
axes[0, 1].set_ylabel('Score')
axes[0, 1].set_ylim(0, 100)
for i, v in enumerate(scores):
    axes[0, 1].text(i, v + 2, str(v), ha='center', fontweight='bold')

# Plot 3: Scatter plot
np.random.seed(42)
x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)
axes[1, 0].scatter(x_scatter, y_scatter, c=np.sqrt(x_scatter**2 + y_scatter**2), 
                   cmap='viridis', s=50, alpha=0.6, edgecolor='black')
axes[1, 0].set_title('Scatter Plot: Random Distribution', fontweight='bold')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')

# Plot 4: Histogram
data = np.random.normal(100, 15, 1000)
axes[1, 1].hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[1, 1].axvline(np.mean(data), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(data):.1f}')
axes[1, 1].set_title('Histogram: Normal Distribution', fontweight='bold')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].legend()

plt.tight_layout()
plt.show()

print("‚úì Matplotlib working correctly!")
print("‚úì All 4 visualization types created successfully!")
````

### Expected Output:
You should see 4 different plots:
1. **Line Plot:** Smooth sine wave
2. **Bar Chart:** Subject scores with labeled values
3. **Scatter Plot:** Random points colored by distance
4. **Histogram:** Distribution with mean line

**What this tests:**
- Plot creation and customization
- Multiple plot types
- Color maps and styling
- Labels and legends

## ü§ñ Test 5: Scikit-learn Machine Learning

Test Scikit-learn with a complete ML pipeline.

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

print("\n" + "=" * 60)
print("TEST 5: SCIKIT-LEARN ML PIPELINE")
print("=" * 60 + "\n")

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

print(f"Dataset: Iris")
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Number of classes: {len(target_names)}")
print(f"Classes: {target_names}")

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nTrain set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")

# Train model
print("\nTraining Random Forest Classifier...")
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)
print("‚úì Model trained successfully!")

# Make predictions
y_pred = model.predict(X_test)

# Evaluate
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:.2%}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(f"\nConfusion Matrix:\n{cm}")

# Classification Report
print(f"\nClassification Report:\n{classification_report(y_test, y_pred, target_names=target_names)}")

# Feature Importance
feature_names = iris.feature_names
importances = model.feature_importances_
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values('Importance', ascending=False)

print("Feature Importance:")
print(importance_df.to_string(index=False))

print("\n‚úì Scikit-learn working correctly!")
print("‚úì Complete ML pipeline executed successfully!")
````

### Expected Output:
You should see:
- Dataset information (Iris: 150 samples, 4 features, 3 classes)
- Train/test split sizes
- Model accuracy (typically 95%+)
- Confusion matrix
- Classification report with precision, recall, F1-score
- Feature importance rankings

**What this tests:**
- Dataset loading
- Data splitting
- Model training
- Predictions
- Performance evaluation
- Feature analysis
