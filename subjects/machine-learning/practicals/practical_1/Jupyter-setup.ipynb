## 1. Python Version Check

```python
import sys
print("Python Version:", sys.version)
print("Python Executable:", sys.executable)

# Check if version is 3.9 or higher
version = sys.version_info
if version.major >= 3 and version.minor >= 9:
    print("âœ“ Python version is compatible (3.9+)")
else:
    print("âœ— Python version should be 3.9 or higher")
```

## 2. Essential Library Imports

```python
libraries = ['numpy', 'pandas', 'matplotlib', 'sklearn']
installed_libs = []

for lib in libraries:
    try:
        __import__(lib)
        version = __import__(lib).__version__
        print(f"âœ“ {lib} {version} - Successfully imported")
        installed_libs.append(lib)
    except ImportError as e:
        print(f"âœ— {lib} - Import failed: {e}")

print(f"\nSuccessfully imported {len(installed_libs)}/{len(libraries)} libraries")
if len(installed_libs) == len(libraries):
    print("âœ“ All essential ML libraries are installed!")
else:
    print("âš  Some libraries are missing. Please check installation.")
```

## 3. Basic Functionality Tests

```python
try:
    import numpy as np
    # Test NumPy
    arr = np.array([1, 2, 3, 4, 5])
    print(f"âœ“ NumPy array: {arr}")
    print(f"âœ“ NumPy mean: {np.mean(arr)}")

    import pandas as pd
    # Test Pandas
    df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
    print(f"âœ“ Pandas DataFrame:\n{df}")

    import matplotlib.pyplot as plt
    # Test Matplotlib
    plt.figure(figsize=(6, 4))
    plt.plot([1, 2, 3], [1, 4, 2])
    plt.title('Test Plot')
    plt.savefig('test_plot.png')
    print("âœ“ Matplotlib plot saved as 'test_plot.png'")

    from sklearn.linear_model import LinearRegression
    # Test Scikit-learn
    X = [[1], [2], [3]]
    y = [1, 2, 3]
    model = LinearRegression()
    model.fit(X, y)
    print(f"âœ“ Scikit-learn LinearRegression score: {model.score(X, y):.2f}")

    print("\nðŸŽ‰ All basic functionality tests passed!")

except Exception as e:
    print(f"âœ— Error in functionality test: {e}")
```
